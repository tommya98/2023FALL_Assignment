# Telco Customer Churn Dataset

(1) Demographic Information

gender: Whether the customer is a female or a male (Female, Male)<br>
SeniorCitizen: Whether the customer is a senior citizen or not (0, 1)<br>
Partner: Whether the customer has a partner or not (Yes, No)<br>
Dependents: Whether the customer has dependents or not (Yes, No)<br>

(2) Customer Account Information

tenure: Number of months the customer has stayed with the company (Multiple different numeric values)<br>
Contract: Indicates the customer’s current contract type (Month-to-Month, One year, Two year)<br>
PaperlessBilling: Whether the customer has paperless billing or not (Yes, No)<br>
PaymentMethod: The customer’s payment method (Electronic check, Mailed check, Bank transfer (automatic), Credit Card (automatic))<br>
MontlyCharges: The amount charged to the customer monthly (Multiple different numeric values)<br>
TotalCharges: The total amount charged to the customer (Multiple different numeric values)<br>

(3) Service Information

PhoneService: Whether the customer has a phone service or not (Yes, No)<br>
MultipleLines: Whether the customer has multiple lines or not (No phone service, No, Yes)<br>
InternetServices: Whether the customer is subscribed to Internet service with the company (DSL, Fiber optic, No)<br>
OnlineSecurity: Whether the customer has online security or not (No internet service, No, Yes)<br>
OnlineBackup: Whether the customer has online backup or not (No internet service, No, Yes)<br>
DeviceProtection: Whether the customer has device protection or not (No internet service, No, Yes)<br>
TechSupport: Whether the customer has tech support or not (No internet service, No, Yes)<br>
StreamingTV: Whether the customer has streaming TV or not (No internet service, No, Yes)<br>
StreamingMovies: Whether the customer has streaming movies or not (No internet service, No, Yes)

- [Kaggle](https://www.kaggle.com/datasets/blastchar/telco-customer-churn)이나 온라인 검색을 통해 자료를 내려받는다.

## 자료 정리

1. 자료를 살펴 전처리가 필요한 부분을 확인하고 어떻게 처리할지 설명하라.
2. 처리내용과 그 이유를 간단히 설명한다.

## EDA

1. 변수들을 정성변수와 정량변수로 나누고, 개별 변수들의 분포르 살펴본다.
2. 변환이 필요하거나 문제가 있는 변수들이 있다면 그 이유를 설명한다.
3. 정성변수를 포함하는 상관관계표를 만든다.
3. 상관관계를 이용하여 변수 선택에 주의해야할 점이 있는지 설명한다.

## Variation Inflation Factor

1. 각 변수들의 variation inflation factor를 구하고 어떤 특징이 있는지 설명하라.
2. 위 정보를 이용하여 삭제할 변수가 있는지 설명하라.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif = pd.Series(vif, index=x.columns, name='VIF')
vif.sort_values(by='VIF', ascending=False)
```

## 훈련과 평가

1. Training set와 test set으로 나눈다. test_size=.4, random_state=0을 사용한다.
2. 필요한 전처리과정을 설명하고 training dataset에 적용한다.
3. 적합한 algorithm이나 다수의 algorithm을 선택하여 훈련시킨다.
4. 모형의 선택기준으로 어떤 scoring을 사용하는 것이 이런 종류의 분석에 적합한지 설명하라.
5. 최종적으로 선택한 모형을 이용하여 test dataset을 예측하고 confusion matrix와 average precision을 보고한다.

데이터를 살펴보았을 때 치우처진 데이터는 많이 보이지 않았다.</br>
다만 정량변수들이 right-skewed된 경향이 조금 있고 tenure가 70이상으로 매우 큰 부분이 일부 있다.</br>
그러나 우선은 내버려 두고 모형을 돌려보려고 한다.

그리고 상관관계를 보니 streamingTV와 streamingMovies 처럼 internetServiece와 관련된 feature에서 상관 관계가 일부 보여 주의해야 한다.

다음으로는 vif를 통해 변수의 다중공산성 여부를 확인한다.

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
vif = pd.Series(vif, index=data.columns, name='VIF')
vif.sort_values(ascending=False, inplace=True)
vif
````

아무래도 정량 변수인 tenure, MonthlyCharges가 크면  TotalCharges가 크게 나올 경향이 높아서 다중공산성도 높게 나온 것 같다.</br>

````python
from sklearn.model_selection import train_test_split

X = data.drop(columns = ['Churn'])
y = data['Churn'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.40, random_state = 0, stratify=y)
````

이대 정량 변수의 스케일이 많이 차이나서 조절을 해준다.

````python
from sklearn.preprocessing import StandardScaler

num_cols = ["tenure", 'MonthlyCharges', 'TotalCharges']
df_std = pd.DataFrame(StandardScaler().fit_transform(data[num_cols].astype('float64')),
                       columns=num_cols)
````

````python
scaler= StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
````

LogisticRegression을 사용해서 분석을 한다.

````python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
lr_score = lr_model.score(X_test,y_test)
lr_score
````

````python
from sklearn.metrics import confusion_matrix, roc_curve, classification_report

lr_pred= lr_model.predict(X_test)
report = classification_report(y_test,lr_pred)
print(report)
````

````python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, lr_pred), annot=True, fmt = "d", linecolor="k")
plt.show()
````

우리가 사용하는 데이터는 탈퇴하지 않을 사람을 잘못 예측하는 것보다 탈퇴할 사람을 잘못 예측하는 것이 더 안좋은 결과이다.</br>
따라서 precision보다 recall이 중요하다.</br>
따라서 roc 커브를 그려보았다.

````python
y_pred_prob = lr_model.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot([0, 1], [0, 1], 'k--' )
plt.plot(fpr, tpr, label='Logistic Regression',color = "r")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve',fontsize=16)
plt.show();
````
